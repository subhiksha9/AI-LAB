```python
import random
sr=[[0,0]]
sl=[[3,3]]
c=1
print("initially at left:",[3,3,1])
print("initially at right:",[0,0,0])
while([3,3] not in sr):
  while c==1:
    slp=sl[::-1]
    srp=sr[::-1]
    trm=random.randint(0,3)
    trc=random.randint(0,3)
    if trm+trc<=2 and slp[0][0]-trm>=0 and slp[0][1]-trc>=0 and (trm!=0 or trc!=0):
      if slp[0][0]-trm-slp[0][1]-trc>0 or (slp[0][0]-trm==0 or trm==0):
        sl.append([slp[0][0]-trm,slp[0][1]-trc])
        sr.append([srp[0][0]+trm,srp[0][1]+trc])
        print("->at left,right:",([slp[0][0]-trm,slp[0][1]-trc,0],[srp[0][0]+trm,srp[0][1]+trc,1]))
        c=0
    else:
        c=1
  while c==0:
    srp=sr[::-1]
    slp=sl[::-1]
    tlm=random.randint(0,3)
    tlc=random.randint(0,3)
    if tlm+tlc<=2 and srp[0][0]-tlm>=0 and srp[0][1]-tlc>=0 and (tlm!=0 or tlc!=0):
      if srp[0][0]-tlm-srp[0][1]-tlc>0 or (srp[0][0]-tlm==0 or tlm==0):
        sr.append([srp[0][0]-tlm,srp[0][1]-tlc])
        sl.append([slp[0][0]+tlm,slp[0][1]+tlc])
        print("->at right,left:",([srp[0][0]-tlm,srp[0][1]-tlc,0],[slp[0][0]+tlm,slp[0][1]+tlc,1]))
        c=1
    else:
        c=0
    if [3,3] in sr:
      break
```
